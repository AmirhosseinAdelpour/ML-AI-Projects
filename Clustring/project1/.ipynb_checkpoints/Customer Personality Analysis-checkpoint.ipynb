## Importing libraries and reading data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Data cleaning

```python
data.info()
```

### Feature engineering & encoding

```python
data["Age"] = 2014 - data["Year_Birth"]
# "Year_Birth" Kidhome 	Teenhome MntWines 	MntFruits 	MntMeatProducts 	MntFishProducts 	
# MntSweetProducts 	MntGoldProds Dt_Customer Recency
```

#### --------------------------------------------------

```python
data.Education.unique()
```

```python
data['Education'] = data['Education'].replace({'Basic':0, '2n Cycle': 1, 'Graduation': 2, 'Master':3, 'PhD':4})
```

#### --------------------------------------------------

```python
data["Marital_Status"].unique()
```

```python
data["Marital_Status"] = data["Marital_Status"].replace({"Single":0, 'Together':1, "Married":1,
                                                         "Divorced":0, "Widow":0, "Alone":0, "Absurd":0, "YOLO":0})
```

#### --------------------------------------------------

```python
data["Children"] = data["Kidhome"] + data["Teenhome"]
```

#### ------------------------------------------------

```python
from datetime import date
last_date = date(2014,10, 4)
data['Seniority'] = pd.to_datetime(data['Dt_Customer'], dayfirst=True)
data['Seniority'] = pd.to_numeric(data['Seniority'].dt.date.apply(lambda x: (last_date - x)).dt.days, downcast='integer')/30
```

#### ----------------------------------------------

```python
data['Spending'] = data['MntWines']+data['MntFruits']+data['MntMeatProducts']+data['MntFishProducts']+data['MntSweetProducts']+data['MntGoldProds']
```

```python
data=data.rename(columns={'MntWines': "Wines",'MntFruits':'Fruits','MntMeatProducts':'Meat','MntFishProducts':'Fish','MntSweetProducts':'Sweets','MntGoldProds':'Gold'})
```

```python
data = data[['Age', 'Education', 'Marital_Status', 'Children', 'Income', 'Spending', 'Seniority', 
            'Wines', 'Fruits', 'Meat', 'Fish', 'Sweets', 'Gold', 'NumDealsPurchases', 'NumWebPurchases',
             'NumCatalogPurchases', 'NumStorePurchases', 'Complain']]
```

```python
data
```

### Removing outliers

```python
data.dropna(inplace=True)
```

```python
data.describe()
```

```python
data = data[data["Income"] < 600000]
```

```python
data
```

### Feature scaling

```python
from sklearn.preprocessing import StandardScaler, normalize
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

# Processing

## Training K-MEANS model

```python
from sklearn.cluster import KMeans
```

### Elbow method to find the optimal number of clusters

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

according to the chart, 5 is the optimal number for clusters.

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(scaled_data)
```

```python
kmeans_data = data
kmeans_data["cluster"] = kmeans.labels_
```

## Training the HIERARCHICAL model

```python
from sklearn.cluster import AgglomerativeClustering
```

### Using the dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(scaled_data, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(scaled_data)
```

```python
hc_data = data
hc_data["cluster"] = hc.labels_
```
