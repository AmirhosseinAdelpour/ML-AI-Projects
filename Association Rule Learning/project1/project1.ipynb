```python
data = pd.read_csv("Market_Basket_Optimisation.csv", header= None)
```

```python
data
```

<div class="alert alert-block alert-info">Making data set ready for Apriori module.</div>

```python
items = []
for i in range(len(data)):
    items.append([str(data.values[i, j]) for j in range(len(data.columns))])
```

# Processing

## Training the Apriori model on the dataset

```python
from apyori import apriori
rules = apriori(transactions = items, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 3, max_length = 4)
```

```python
results = list(rules)
```

```python
results
```

# Displaying results

```python
# gathering items in a list
rules_list = []
for i in range(len(results)):
    rules_list.append([j for j in results[i][0]])
```

```python
# gathering item's confidences 
rules_confidences = []

for i in range(len(results)):
    rules_confidences.append([results[i][2][0][2]])
```

```python
list(zip(rules_list, rules_confidences))
```
