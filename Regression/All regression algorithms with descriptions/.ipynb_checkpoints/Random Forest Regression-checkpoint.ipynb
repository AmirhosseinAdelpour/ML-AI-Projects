# Random Forest Regression

## Descriptions

This algorthm which is related to decision tree, is a algorith that:
1-Takes n random points of our data and makes a decision tree by them.
2-Then it takes another n random points from our data and makes another decision tree by them and will continue this as many times as we want.
3-Finally it predict our point in all the decision trees that has been made and will get the average of them for our answer.

<div class="alert alert-block alert-info">For removing outliers, we have two methods.<br>
IQR method and STD method.<br><br>
<b>In IQR method</b> first we find the first and third quarters and then for finding the interquartile range(IQR) we subtract first quarter from third quarter.<br>Then all the points upper than Q3 + (1.5 * IQR) and all the points lower than Q1 - (1.5 * IQR) will be outliers.

<b>In STD method</b>, first we find the mean and standard deviation of the data. then we triple the std.<br>
we subtract triple std from mean(lower) and sum triple std with mean(upper) and now all the data points lower than lower and upper than upper are outliers.</div>

```python
# plot all columns to see if there are outliers or not

col = data.columns

fig, axs = plt.subplots(5, 3, figsize=(18, 18))

x = 0
y = -1
counter = 0

for i in col:
    if counter < 5: 
        if y <= 1:
            y += 1
            axs[x][y].scatter(data[i], data.price)
            axs[x][y].set_title(i, fontsize=15)
        else:
            y = 0
            x += 1
            counter += 1
            axs[x][y].scatter(data[i], data.price)
            axs[x][y].set_title(i, fontsize=15)

plt.show()
```

#### Splitting data into train and test sets

```python
from sklearn.model_selection import train_test_split
```

```python
x = np.asanyarray(data.drop("price", axis=1))
y = np.asanyarray(data.price)
y = y.reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=2)
```

### Processing

#### Creating model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regr = RandomForestRegressor()
regr.fit(x_train, y_train)
```

#### Evaluating model

```python
from sklearn.metrics import r2_score
```

```python
y_pred = regr.predict(x_test)
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

<div class="alert alert-block alert-info">In addition to the r2, we have an adjusted r2 too.<br>R2 value always increases by adding new variables although that variables are completely irrelevant.<br>So for fixing this problem we use adjusted_r2 which shows us the true value and will always be equal or lower than r2.<div>

```python
# Adjusted R2
1 - (((1-r2)*(len(y_pred)-1)) / (len(y_pred)-x.shape[1]-1))
```
